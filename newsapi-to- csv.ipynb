```python
section='general_news'
ENDPOINT=f"https://newsapi.org/v2/everything" # f string has the url with the option to change variables inside if needed. 
PARAMETERS = {'q':'bitcoin', # q is the query and we have requested info about bitcoin. we could use anything other than bitcoin
              'apikey': News_Api_Key # this is my api keys hidden. 
             }

res=requests.get(url = ENDPOINT, params = PARAMETERS) # requesting thr data from the server and storing it to 'res'
lim=res.json() # turning the data into json file 
```
